 # Performing Exploratory Data Analysis (EDA)

```python
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
df = pd.read_csv('/content/train.csv')
```

```python
# obtaining the first 5 columns of the dataset to understand the data
df.head()
```

```python
# obtaining the different data types
df.dtypes
```

```python
#obtaining the summary statistics
df.describe()
```

```python
# finding the number of rows and columns
df.shape
```

Handling the missing values

```python
print(df.isnull())
```

```python
# to find the total number of missing values
print(df.isnull().sum)
```

Observation: There a few missing values in age and cabin

```python
# filling the missing values in age column
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# filling missing values in cabin column
df.drop(columns='Cabin', inplace=True)
```

Converting categorical variables to numerical variables

```python
# Convert 'Sex' column to numerical
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Convert 'Embarked' column to numerical
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

Exploring the relationships between the variables using heat maps from seaborn package

```python
# Correlation matrix
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numeric_columns].corr()
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Visualising patterns and trends among the variables

```python
# histogram for ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# bar plot for survival rates by sex
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rates by Sex')
plt.xlabel('Sex (0=Male, 1=Female)')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# bar plot of survival rates by class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rates by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# bar plot of survival rates by embarked location
plt.figure(figsize=(8, 6))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rates by Embarked Location')
plt.xlabel('Embarked Location (0=C, 1=Q, 2=S)')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# pair plot of numerical features
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass', 'Sex', 'SibSp', 'Parch']])
plt.show()
```

Summary:

Age Distribution: The age distribution shows a right-skewed pattern with most passengers being between 20 to 40 years old.

Survival Rates by Sex: Females had a significantly higher survival rate compared to males.

Survival Rates by Class: Passengers in first class had the highest
survival rate, followed by second class, and then third class.

Survival Rates by Embarked Location: Passengers who embarked from location 'C' had the highest survival rate.
